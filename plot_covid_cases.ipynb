```python
import numpy as np # Load the numpy library
import matplotlib.pyplot as plt
from matplotlib.dates import date2num, num2date, datetime
```

```python
cases = np.loadtxt('Casos_Diarios_Estado_Nacional_Confirmados_20200709.csv', delimiter=',') # Loads cases file
time = np.loadtxt('Fechas.csv', delimiter=',')
```

```python
time = np.reshape(time, (np.int(time.shape[0]/3), 3)) # Why this
```

```python
# Loads time mm,dd,yyyy HH,mm
day = time[:,0]
month = time[:,1]
year = time[:,2]
time_final = np.zeros(year.shape[0])
for ii in range(year.shape[0]):
    time_final[ii] = date2num(datetime.datetime(np.int(year[ii]), np.int(month[ii]),
                         np.int(day[ii])))
```

```python
cases = np.ma.masked_invalid(cases) # Get rid of invalid values
```

```python
fig, ax = plt.subplots(1, 1, sharex=True, figsize=(14, 6))

tit = 'National Number of COVID cases'
units = 'Number of cases per day'
ax.plot(num2date(time_final), cases, lw=1, color='m')
ax.bar(time_final, cases)
ax.set_title(tit)
ax.set_ylabel(units)
# ax.set_ylim(0, 125)

fi1 = date2num(datetime.datetime.strptime('2019-06', '%Y-%m'))
fi2 = date2num(datetime.datetime.strptime('2020-01', '%Y-%m'))
#ax.set_xlim(fi1, fi2)
# plt.savefig('Jair_2019.png', bbox_inches='tight')
```
